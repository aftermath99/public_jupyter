```python
# import the used libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

pd.__version__
```

```python
# Build base match data dataframe - originally read from BQ
match_data = pd.DataFrame()

competitionid = [4 , 4, 4]
competitionname = ['World Cup', 'World Cup', 'World Cup']
matchid = ['311168', '311137', '311154']
team_name = ['Czech Republic', 'Ghana', 'Italy']
teamid = [367, 1219, 119]
op_team_name = ['USA', 'USA', 'USA']
op_teamid = [596, 596, 596]

match_data['competitionid'] = competitionid
match_data['competitionname'] = competitionname
match_data['matchid'] = matchid
match_data['team_name'] = team_name
match_data['teamid'] = teamid
match_data['op_team_name'] = op_team_name
match_data['op_teamid'] = op_teamid
match_data
```

```python
# Goal data, keyed by matchid and teamid
goal_data = pd.DataFrame()

matchid = ['311168', '311168', '311137', '311137', '311154', '311154']
teamid = [367, 596, 1219, 596, 119, 596]
goals = [3, 1, 1, 2, 4, 0]

goal_data['matchid'] = matchid
goal_data['teamid'] = teamid
goal_data['goals'] = goals
goal_data
```

```python
# Now, merge the goal data to the match data, once for the home team and again for the away
merged_df = pd.merge(match_data, goal_data, 
                     left_on = ['matchid', 'teamid'], 
                     right_on = ['matchid', 'teamid'], 
                     how='left')

# Merging renames matching columns to make explicit which columns belonged to which dataframes
merged_df = merged_df.rename(columns={"teamid_x":"teamid", "goals":"home_goals"})

merged_df = pd.merge(merged_df, goal_data, 
                     left_on = ['matchid', 'op_teamid'], 
                     right_on = ['matchid', 'teamid'], 
                     how='left')
merged_df = merged_df.rename(columns={"teamid_x":"teamid", "goals":"op_goals"})

# Drop duplicate columns
merged_df.drop("teamid_y", axis=1, inplace=True)
merged_df
```
